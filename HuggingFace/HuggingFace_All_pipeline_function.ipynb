# **Hugging Face **

**NLP:** enables computer to generate human language

**LLM:** characterized by their massive size, extensive training data, and ability to perform a wide range of language tasks with minimal task-specific training. GPT,LLAMA


**Pipeline function**

returns end to end object, performs nlp tasks on one or several texts

pre processing -> models -> post processing

* text classification
* zero-shot classification
* text generation
* text completion(mask filling)
* Token Classifiaction
* Question Answering
* Summarization
* Translation


```python
#Sentiment Analysis (pipeline API)
from transformers import pipeline
classifier=pipeline("sentiment-analysis")
classifier([
    "I'm happy to learn about transformers",
    "I will go to dhaka but i don' like dhaka."
])
```

Zero Shot Pipeline
allows to provide lables for the Classification

```python
from transformers import pipeline
classifier=pipeline("zero-shot-classification")
classifier(
    "This is a course about the transformers library ",
    candidate_labels=["fun","education","business"],#labels
)
```

**Text Generation Pipeline**

input prompt to generate text

```python
from transformers import pipeline
generator=pipeline("text-generation")
generator("when we reached stanford we went to")
```

```python
from transformers import pipeline
generator= pipeline("text-generation", model="distilgpt2")
generator("In this course of Deeplearning we will learn",
          max_length=30,
          num_return_sequences=2,
)
```

**Fill Mask pipeline**

predicts missing words in a sentence

```python
from transformers import pipeline
unmasker=pipeline("fill-mask")
unmasker("This course will teach you all about <mask> model",top_k=2)
```

**Ner pipeline**

identifies entities such as persons organizations or locations

```python
from transformers import pipeline
ner=pipeline("ner",grouped_entities=True)
ner("My Name is Sushmita and i work at Hugging Face in Paris.")
```

**Question Answering pipeline**

extracts answers to a question from a given context

```python
from transformers import pipeline
question_answer=pipeline("question-answering")
question_answer(
    question="Where does he work?",
    context="He is Megh, he works at huggingface in Paris"
)
```

**Translation Pipeline **

translates text from one language to another

```python
from transformers import pipeline
translator=pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en")
translator("Je m'appelle Sushmita, Je suis etudiante")
```
