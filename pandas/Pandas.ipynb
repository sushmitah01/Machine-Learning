# ** Data Cleaning  **


*   Empty Cells
*   Data in wrong format
*   Wrong Data
*   Duplicates



```python
#remove rows
df = pd.read_csv('data.csv')

new_df = df.dropna() #By default, the dropna() method returns a new DataFrame, and will not change the original.

print(new_df.to_string())
```

to change the original DataFrame, use the **inplace = True ** argument:

```python
df=pd.read_csv('data.csv')
df.dropna(inplace = True)
print(df.to_string()) #it removes all null values from original data frame insted of creating a new one
```

to replace empty values with a value use **fillna()** method

```python
df=pd.read_csv('data_csv')
df.fillna(130,inplace=True) #REPLACE NULL VALUES WITH NUMBER  30
```

To only replace empty values for **one column**, specify the column name for the DataFrame:

```python
df=pd.read_csv('data_csv')
df["Calories"].fillna(130,inplace=True) # Replace NULL values in the "Calories" columns with the number 130
```

# Mean, Median, Mode


*   Mean= the average value





```python
df = pd.read_csv('data.csv') # Calculate the MEAN, and replace any empty values with i

x = df["Calories"].mean()

df.fillna({"Calories": x}, inplace=True)
```

Calculate the **MEDIAN** and replace any empty values with it:


*   **Median** = the value in the middle, after you have sorted all values ascending.



```python
df = pd.read_csv('data.csv')

x = df["Calories"].median()

df.fillna({"Calories": x}, inplace=True)
```

Mode = the value that appears most frequently.

```python
df = pd.read_csv('data.csv')

x = df["Calories"].mode()[0] #if theres multiple frequent values it will take the first one [0]

df.fillna({"Calories": x}, inplace=True)
```

# Cleaning Wrong **Format**

```python
#
df = pd.read_csv('data.csv')

df['Date'] = pd.to_datetime(df['Date'], format='mixed')

print(df.to_string())
```

```python
#remove rows with a null value
df.dropna(subset=['Date'], inplace = True)
```

# Fixing Wrong Data

```python
df.loc[7, 'Duration']=45 #row 7, value 45
```

To fix values in **multiple rows**

```python
#loop through them
#for example
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
```

**Delete** Rows

```python
#deleting rows
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
```

# Removing Dupolicates

**To discover duplicates**

```python
print(df.duplicated())#retuns true for every row that is duplicate, otherwise False
```

now **to drop that** duplicates

```python
df.drop_duplicates(inplace = True)
```
