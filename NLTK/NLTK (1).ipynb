# **Lemmatization**
-the process of reducing a word to its base or dictionary form, called a lemma.
-(Helps NLP systems understand that different forms of a word are actually the **same word**.)
-uses **vocabulary** and **part-of-speech** to find the correct base form.

Note::
-better than stemming
such as - running to run better to good

```python
from nltk.stem import WordNetLemmatizer
text="feet cats wolves talked"
words=['Hello', 'world', '!', 'NLTK', 'makes', 'text', 'processing', 'easy', '.']
lemmatizer = WordNetLemmatizer()
for w in words:
  print (lemmatizer.lemmatize(w))
# Verbs
print(lemmatizer.lemmatize("running", pos="v"))  # 'run'
print(lemmatizer.lemmatize("ate", pos="v"))      # 'eat'

# Nouns
print(lemmatizer.lemmatize("cars", pos="n"))     # 'car'
print(lemmatizer.lemmatize("geese", pos="n"))    # 'goose'

# Adjectives
print(lemmatizer.lemmatize("better", pos="a"))   # 'good'
```

**Used in  **

*   Text classification

*   Information retrieval

*   Sentiment analysis

```python
#using ml algorithms
# Convert text to features (BoW, TF-IDF).
# Train a classifier (Naive Bayes, SVM, Logistic Regression).

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

texts = ["I love this movie", "This movie is terrible"]
labels = [1, 0]  # 1 = positive, 0 = negative

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

model = MultinomialNB()
model.fit(X, labels)

test = vectorizer.transform(["I hate this movie"])
print(model.predict(test))  # [0] → Negative
```

# **Normalization**
standard text
(happens all of the above here)

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

text = "The cats are running on the mats!!!"

# 1. Lowercase
text = text.lower()

# 2. Remove punctuation
text = re.sub(r'[^\w\s]', '', text)

# 3. Tokenize
tokens = word_tokenize(text)

# 4. Remove stop words
stop_words = set(stopwords.words('english'))
tokens = [w for w in tokens if w not in stop_words]

# 5. Lemmatization
lemmatizer = WordNetLemmatizer()
tokens = [lemmatizer.lemmatize(w) for w in tokens]

print(tokens)
```

# NER(named entity recognition)
only works in english
**NER tells you what it is in the real world.**

---
🔹 Types of Named Entities (in NLTK)

PERSON → People’s names

ORGANIZATION → Companies, institutions, agencies

GPE → Geo-political entities (countries, cities, states)

LOCATION → Non-political locations (mountains, rivers, landmarks)

DATE → Dates (years, months, etc.)

TIME → Times of day

MONEY → Money values

PERCENT → Percentages

FACILITY → Buildings, airports, highways, etc.



```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

sentence="Barack Obama was the 44th President of the United States."
tokens = word_tokenize(sentence)
tags = pos_tag(tokens)
tree = ne_chunk(tags)
print(tree)
```

**extracting named entities as a text**

```python
entities = []
for chunk in ne_chunk(tags):
    if hasattr(chunk, 'label'):  # means it’s a named entity
        entity = " ".join(c[0] for c in chunk)
        entities.append((entity, chunk.label()))

print(entities)
```

```python
# with multiple entities

sentence = "Apple is planning to open a new office in New York in April 2025 with an investment of $5 million."
tokens = word_tokenize(sentence)
tags = pos_tag(tokens)
tree = ne_chunk(tags)

entities = []
for chunk in tree:
    if hasattr(chunk, 'label'):
        entity = " ".join(c[0] for c in chunk)
        entities.append((entity, chunk.label()))

print(entities)
```

**1.Frequency count**

```python
#frequnecy count
#useful for
#------------------------------------------------------------
#        Keyword extraction
#        Data exploration
#        Finding stopwords (common but unhelpful words)

import nltk
from nltk import FreqDist
from nltk.tokenize import word_tokenize

text = "I love NLP and I love Python. Python is amazing for NLP."
tokens = word_tokenize(text)

fdist = FreqDist(tokens)
print(fdist.most_common(5))   # top 5 words
print(fdist['Python'])        # frequency of 'Python'

fdist.plot(10)   # frequency plot of top 10 words
```

**2.Collection & N-grams**
---------------------------------
Collocation = words that often appear together more than by chance.

Example: "New York", "machine learning".

N-gram = sequence of n words (bigrams = 2, trigrams = 3)


```python
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
from nltk.util import ngrams

text = "I love machine learning and deep learning with Python"
tokens = word_tokenize(text)

# Bigrams
finder = BigramCollocationFinder.from_words(tokens)
print(finder.nbest(BigramAssocMeasures.likelihood_ratio, 3))  # top 3 bigrams

# Trigrams using ngrams()
trigrams = list(ngrams(tokens, 3))
print(trigrams)

# ------------------------------------------------------------

# Useful for:

# Identifying key phrases

# Improving search engines

# Language modeling
```

**3. Concordance & Similar Words**
--------------------------------
Concordance → shows where a word occurs in context.

Similar words → finds words used in similar contexts.

```python
from nltk.book import text1   # "Moby Dick" (comes with NLTK)

text1.concordance("whale")       # shows contexts of "whale"
text1.similar("ship")            # words used in similar places as "ship"
text1.common_contexts(["man", "woman"])  # contexts where both appear

# --------------------------------------------
#  Useful for:

# Linguistic analysis

# Understanding word meaning by context
```

**4. Text Classification: Naive Bayes (NLTK)**
---

Naive Bayes Classifier = probabilistic model for text classification.

Commonly used for spam detection, sentiment analysis, topic classification.

Steps:

Extract features from text (usually bag-of-words).

Train classifier.

Predict new sample

```python
import nltk

# Training data (toy example)
train_data = [
    ({"word": "love", "sentiment": True}, "pos"),
    ({"word": "great", "sentiment": True}, "pos"),
    ({"word": "bad", "sentiment": False}, "neg"),
    ({"word": "terrible", "sentiment": False}, "neg"),
]

# Feature extractor
def word_features(words):
    return {word: True for word in words.split()}

# Updated training set
train_set = [
    (word_features("I love this movie"), "pos"),
    (word_features("This is a great film"), "pos"),
    (word_features("This movie is terrible"), "neg"),
    (word_features("I hate this film"), "neg"),
]

# Train
classifier = nltk.NaiveBayesClassifier.train(train_set)

# Test
print(classifier.classify(word_features("I love it")))
print(classifier.classify(word_features("This is bad")))
classifier.show_most_informative_features()
```
